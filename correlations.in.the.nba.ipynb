## Win Percentage per Minute Correlations in the NBA

### Finding the three most correlated minutes

```python
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline 

#open datafile and create new list for data in the file
data_file = open("nba.csv")
ppm_list = []
for a in data_file:
    ppm_list.append(a);

#get rid of headers
del ppm_list[0]

#create new list for values 
new_list = []
for j in range(1,48):
    new_list.append([])

#create a list within a list to give the win percentage per minute for each team for every minute
for b in range(len(ppm_list)):
    ppm_list[b] = ppm_list[b].split(",");
    for i in range(0,47):
        new_list[i].append(float(ppm_list[b][i+1]));
        
#create new list for anchor variable in the 48th minute
fe_list = []
for c in range(len(ppm_list)):
    fe_list.append(float(ppm_list[c][49]));

#perform pearson r test correlating all the minutes to the 48th minute 
for d in range(0,47):
    scipy.stats.pearsonr(new_list[d], fe_list)

print("47th minute pearson value and p value", scipy.stats.pearsonr(new_list[46], fe_list))
print("45th minute pearson value and p value", scipy.stats.pearsonr(new_list[44], fe_list))
print("46th minute pearson value and p value", scipy.stats.pearsonr(new_list[45], fe_list))
    
    
```

In this code I used "for" loops to put the data in organized lists to correlate using the pearson r test.  I put the data from the win percentage per minute for every minute into a list and then compared it to the data from the final 48th minute and looked for correlation.  The closest value to 1 or negative 1 will give me the most correlated values and I disregaurded the p values since there are not enough data points in the data sets.  There was the most correlation in the 47th minute, then the 45th, then the 46th.

### Making scatter plots for three most correlated minutes

```python
#plot a scatter plot of the most correlated minute
plt.scatter(new_list[46], fe_list)
plt.title("Correlation of 47th minute and 48th minute")
plt.xlabel("47th minute")
plt.ylabel("48th minute")
```

```python
#plot a scatter plot of the second most correlated minute
plt.scatter(new_list[44], fe_list)
plt.title("Correlation of 45th minute and 48th minute")
plt.xlabel("45th minute")
plt.ylabel("48th minute")
```

```python
#plot a scatter plot of the third most correlated minute
plt.scatter(new_list[45], fe_list)
plt.title("Correlation of 46th minute and 48th minute")
plt.xlabel("46th minute")
plt.ylabel("48th minute")
```

I made scatter plots for each of the three most correlated minutes.

In conclusion, the minutes where the win percentage in the minute is most correlated to the actual win percentage of the game come in the final minutes.  This makes sense because as a game comes closer to the end the win percentage per minute should be closer to the actual outcome of the game.  Especially in basketball where so many points are scored, there is not much that can happen in the last minute of a game to change the outcome if it is more than a 3 point game.
